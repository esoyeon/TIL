```python
import pandas as pd 
import os

os.listdir()
os.chdir('./mulcam_team3/')
```

```python
df = pd.read_csv('pre_processed_naver_interest_article.csv')
df 
```

```python
text = df['article'][0]
text
```

```python
from ekonlpy.sentiment import MPCK
mpck = MPCK()
tokens = mpck.tokenize(text)
ngrams = mpck.ngramize(tokens)
score = mpck.classify(tokens + ngrams, intensity_cutoff=1.5)
```
