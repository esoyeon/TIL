```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Make_classification

- n_samples : data의 수, 기본값은 100  
- n_features : feature의 수, 기본값은 20  
- n_informative : 독립변수중 종속변수와 상관관계가 있는 성분의 수(y와 관련이 있는 x의 수), 기본값은 2  
- n_redundant : 다른 독립변수들의 선형조합으로 만들어지는 독립변수의 수(다른 x들의 선형조합으로 만들어 지는 x의 수), 기본값은 2  
- n_classes : 종속변수의 class 수, 기본값은 2 (이진 데이터)  
- n_clusters_per_class : 각 클래스당 클러스터 수, 기본값은 2, 클래스당 클러스터 수를 늘리게 되면 클래스 끼리 잘 분리되지 않은 상태의 데이터를 생성하게 된다., 주의 사항으로 n_classes * n_clusters_per_class는 2 ** n_informative보다 작거나 같게 설정해야 한다.  
- random_state : 난수 발생, 난수값을 다르게하면 각자 다른 형태의 데이터가 나오게 된다. 

```
from sklearn.datasets import make_classification
```

```
X, y = make_classification(n_samples = 100, n_features = 1, n_redundant= 0, 
                           n_informative = 1, n_clusters_per_class = 1,
                           random_state = 30)
```

```
X, y
```

```
X.shape, y.shape
```

```
plt.figure(figsize = (10,5))
plt.scatter(X, y, c=y, marker = '*', s=200, edgecolors='k') # k = black
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```
# 독립변수, 즉 n_features를 2로 설정한 데이터를 그래프로 그리기
X, y = make_classification(n_samples = 100, n_features = 2, n_redundant=0,
                           n_informative =1, n_clusters_per_class =1,
                           random_state = 30)
X.shape, y.shape
```

```
X, y
```

```
plt.figure(figsize=(10,5))
plt.scatter(X[:,0], X[:,1], marker = '*', c = y, edgecolors='k', s=200)
plt.title('2개의 독립변수를 가진 데이터')
plt.xlabel('X의 첫번째 변수')
plt.ylabel('X의 두번째 변수')
plt.colorbar()
plt.show()
```

### Logistic Regression  기본

```
from sklearn.linear_model import LogisticRegression
```

```
log_reg = LogisticRegression()
```

```
log_reg.fit(X,y)
```

```
# Prediction과 y값 10개만 비교해보기
log_reg.predict(X)[:10], y[:10]
```

```
log_reg.coef_, log_reg.intercept_
```

```
log_reg.decision_function(X)
# 양수면 1, 음수면 0으로 보자
```

```
log_reg.predict(X) #decision function의 결과
```

```
log_reg.predict_log_proba(X)[:10] #로지스틱은 확률을 구하니까, 확률은 얼마?
```

```
# Dicision Boundary 그리기
# wx+b = 0일때, sigmoid 함수가 0.5 !
# 즉, Wx+b = 0인 지점이 decision boundary
# WX = 0인 지점 => W1X1 + W2X2 + W0 = 0

X_1 = np.linspace(X[:,0].min(), X[:, 0].max()) # 50개가 default
W = log_reg.coef_.flatten()
W1 = W[0]
W2 = W[1]
X_2 = -(W1*X_1 + log_reg.intercept_) / W2
X_2
```

```
plt.figure(figsize = (10,5))
plt.scatter(X[:,0], X[:,1], c = y, marker = '*', edgecolors='k', s =200)
plt.xlabel("X의 첫 번째 독립변수")
plt.ylabel("X의 두 번째 독립변수")
 
#Decision Boundary 
plt.plot(X_1, X_2)

plt.show()
```

#### Performance Measure

```
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```
print(accuracy_score(y, log_reg.predict(X)))
print(precision_score(y, log_reg.predict(X)))
print(recall_score(y, log_reg.predict(X)))
print(f1_score(y, log_reg.predict(X)))
```

```
log_reg.score(X, y) 
# Logistic Regression 내부적으로는 Accuracy가 Default
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y, log_reg.predict(X))
```

```
from sklearn.metrics import classification_report
print(classification_report(y, log_reg.predict(X)))
```

### 규제값에 따른 변화 

#### C값 조절 

```
from sklearn.model_selection import train_test_split
X, y = make_classification(n_samples = 600, n_features=50, n_redundant = 0,
                          n_informative = 1, n_clusters_per_class = 1,
                          random_state = 30)
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

for C in [0.01, 0.1, 1, 10, 100]:
  print("C Value: ", C)
  lin_reg = LogisticRegression(penalty='l2', C = C ).fit(X_train, y_train)
  print("train 정확도: ", accuracy_score(y_train, lin_reg.predict(X_train)))
  print("test 정확도: ", accuracy_score(y_test, lin_reg.predict(X_test)))
  print('-'*10)
```

#### 규제값 조절

```
# X에 대해서 LOGISTIC REGRESSION C값(규제)을 바꿔가면서 한 번 작성 해주세요.
# 각 결과에서 Precision, Recall, f1_score, accuracy를 출력해주세요.
# 규제는 l1규제 한 번, l2규제로도 한번 해보세요.

# l1 규제 에러=> solver: liblinear => lbfgs(L-BFGS) 
# l1 규제시 liblinear 사용해야 함. solver='liblinear'로

log_reg = LogisticRegression(penalty='l1')
log_reg.fit(X_train, y_train)
print(accuracy_score(y_test, log_reg.predict(X_test)))
print(precision_score(y_test, log_reg.predict(X_test)))
print(recall_score(y_test, log_reg.predict(X_test)))
print(f1_score(y_test, log_reg.predict(X_test)))
```

```
log_reg = LogisticRegression(penalty='l1', solver = 'liblinear')
log_reg.fit(X_train, y_train)
print(accuracy_score(y_test, log_reg.predict(X_test)))
print(precision_score(y_test, log_reg.predict(X_test)))
print(recall_score(y_test, log_reg.predict(X_test)))
print(f1_score(y_test, log_reg.predict(X_test)))
```
