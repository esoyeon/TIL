```python
text = df['article'][:3]
text
```

```python
from ekonlpy.sentiment import MPCK
mpck = MPCK()
```

```python
ngrams =  []
for i in text:
    tokens = mpck.tokenize(i)
    ngram = mpck.ngramize(tokens)
    
    for j in ngram:
        if j not in ngrams:
            ngrams.append(j)
#     score = mpck.classify(tokens + ngrams, intensity_cutoff=1.5)
```
