```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.01)
y = np.sin(x)
```

```python
plt.figure().add_subplot().set_title('한글 제목 테스트 Plot')
plt.plot(x, y)
```
